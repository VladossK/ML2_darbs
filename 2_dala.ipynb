# Nepiecieams bibliotkas

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Datu sagatavoana

```python
# Atlasm tikai skaitlisks pazmes
X = df[['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day',
        'Physical_Activity_Hours_Per_Day', 'Grades']]

# Normalizjam datus
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Hierarhisk klasterizcija (3 eksperimenti)

```python
# Eksperiments 1  'ward'
linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma  Ward metode")
plt.show()

# Eksperiments 2  'complete'
linked = linkage(X_scaled, method='complete')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma  Complete metode")
plt.show()

# Eksperiments 3  'average'
linked = linkage(X_scaled, method='average')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma  Average metode")
plt.show()
```

# K-vidjo algoritms (KMeans) + Silueta koeficients

```python
# Saglabsim Silueta rezulttus
silhouette_scores = []

for k in range(2, 7):  # 5 dadas K vrtbas
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append((k, score))
    print(f"K={k}, Silueta koeficients: {score:.4f}")

# Attlojam silueta koeficientus
k_vals, scores = zip(*silhouette_scores)
plt.plot(k_vals, scores, marker='o')
plt.title("Silueta koeficients dadiem K")
plt.xlabel("K vrtba")
plt.ylabel("Silueta koeficients")
plt.grid(True)
plt.show()
```

# Vizualizcija ar izvlto K

```python
# Pieemsim K=3 k piemrotko
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X_scaled)
df['Cluster'] = labels

# 2D scatter plot (piemrs)
sns.scatterplot(data=df, x='Sleep_Hours_Per_Day', y='Grades', hue='Cluster', palette='Set1')
plt.title("K-vidjo klasterizcija  Miegs vs Vrtjums")
plt.show()
```
