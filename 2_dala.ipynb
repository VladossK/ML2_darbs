# Nepieciešamās bibliotēkas

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Datu sagatavošana

```python
# Atlasām tikai skaitliskās pazīmes
X = df[['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day',
        'Physical_Activity_Hours_Per_Day', 'Grades']]

# Normalizējam datus
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Hierarhiskā klasterizācija (3 eksperimenti)

```python
# Eksperiments 1 – 'ward'
linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma – Ward metode")
plt.show()

# Eksperiments 2 – 'complete'
linked = linkage(X_scaled, method='complete')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma – Complete metode")
plt.show()

# Eksperiments 3 – 'average'
linked = linkage(X_scaled, method='average')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title("Dendrogramma – Average metode")
plt.show()
```

# K-vidējo algoritms (KMeans) + Silueta koeficients

```python
# Saglabāsim Silueta rezultātus
silhouette_scores = []

for k in range(2, 7):  # 5 dažādas K vērtības
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append((k, score))
    print(f"K={k}, Silueta koeficients: {score:.4f}")

# Attēlojam silueta koeficientus
k_vals, scores = zip(*silhouette_scores)
plt.plot(k_vals, scores, marker='o')
plt.title("Silueta koeficients dažādiem K")
plt.xlabel("K vērtība")
plt.ylabel("Silueta koeficients")
plt.grid(True)
plt.show()
```

# Vizualizācija ar izvēlēto K

```python
# Pieņemsim K=3 kā piemērotāko
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X_scaled)
df['Cluster'] = labels

# 2D scatter plot (piemērs)
sns.scatterplot(data=df, x='Sleep_Hours_Per_Day', y='Grades', hue='Cluster', palette='Set1')
plt.title("K-vidējo klasterizācija – Miegs vs Vērtējums")
plt.show()
```
