# Statistiskie rdtji

```python
# Vidjs vrtbas
means = df.groupby('Stress_Level').mean(numeric_only=True)
print("Vidjs vrtbas:\n", means)

# Dispersijas
variances = df.groupby('Stress_Level').var(numeric_only=True)
print("\nDispersijas:\n", variances)
```

# Datu sagatavoana (3 pazmes + normalizcija)

```python
from sklearn.preprocessing import StandardScaler

# Atlasm 3 piemrotas pazmes klasterizcijai
X = df[['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Grades']]

# Normalizjam datus
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# K-vidjo algoritms + Silueta koeficients (26 klasteri)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

silueta_koeficienti = []

for k in range(2, 7):  # no 2 ldz 6 klasteriem
    modelis = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = modelis.fit_predict(X_scaled)
    sil_score = silhouette_score(X_scaled, labels)
    silueta_koeficienti.append(sil_score)
    print(f"K={k}, Silueta koeficients: {sil_score:.4f}")

# Vizualizcija
plt.plot(range(2, 7), silueta_koeficienti, marker='o')
plt.title("Silueta koeficients dadiem K")
plt.xlabel("Klasteru skaits")
plt.ylabel("Silueta koeficients")
plt.grid(True)
plt.show()

```

# K=3 piemrs: scatter plot

```python
modelis = KMeans(n_clusters=3, random_state=42, n_init='auto')
df['Cluster'] = modelis.fit_predict(X_scaled)

sns.scatterplot(data=df, x='Sleep_Hours_Per_Day', y='Grades', hue='Cluster', palette='Set1')
plt.title("Klasterizcija: Miegs vs Vrtjums")
plt.show()
```

# Hierarhisk klasterizcija (Ward metode + dendrogramma)

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

Z = linkage(X_scaled, method='ward')
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title("Hierarhisk klasterizcija (ward)")
plt.xlabel("Datu objekti")
plt.ylabel("Attlums")
plt.show()
```
